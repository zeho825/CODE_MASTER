# 코딩 테스트 연습 1주차

## 초급
![alt text](image.png)

```python
# 초급 문제
data1 = input("> 문자열을 입력해주세요")
result = ""
for i in data1:
    if i.isupper() == True:
        result += i.lower()
    elif i.islower() == True:
        result += i.upper()
    else:
        result += i
print(f"입력된 문자열 : {data1}")
print(f"변환된 문자열 : {result}")
```

## 중급
![alt text](image-1.png)

```python
# 띄어쓰기를 기준으로 단어들만 순서 변경

data1 = input("> 문자열을 입력해주세요")
data2 = data1.split()
result = ""

for i in data2:
    result += i[::-1] + " "
print(f"입력된 문자열 : {data1}")
print(f"변환된 문자열 : {result}")
```

## 고급
![alt text](image-2.png)

```python
## 고급
data1 = input("> 문자열을 입력해주세요")

resent = ""
store = ""
count = 1

for i in range(1, len(data1)):
    if data1[i-1] == data1[i]:
        resent = data1[i-1]
        count += 1

    elif data1[i-1] != data1[i]:
        store += data1[i-1] + f'{count}'
        resent = data1[i-1]
        count = 1
print(f"입력된 문자열 : {data1}")
print(f"변환된 문자열 : {store+resent}"+f'{count}')
```
