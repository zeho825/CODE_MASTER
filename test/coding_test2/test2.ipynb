## 중급
![alt text](image-1.png)

```python
data1 = input("> 문자열 입력해주세요")

def solve_intermediate(x):
    if not x:
        return 0
    max_len = 1
    current_len = 1
    for i in range(1, len(x)):
        if x[i] == x[i-1]:
            current_len += 1
        else:
            max_len = max(max_len, current_len)
            current_len = 1
    max_len = max(max_len, current_len)
    return max_len

print(data1)
print(solve_intermediate(data1))
```

## 고급
![alt text](image-2.png)

```python
import re

data1 = input("> 문자열 입력해주세요")

def solve_advanced(x):
    numbers = re.findall(r'\d+', x)
    if not numbers:
        return 0
    return max(map(int, numbers))

# Test cases
print(data1)
print(solve_advanced(data1))
```
