```python
data1 = input("> 문자열 입력해주세요")
result = {}

for i in data1:
    if i in result.keys():
        result[i] += 1
    else:
        result[i] = 1

print(result)
max_result = max(result, key=result.get)
print(max_result)
```

```python
my_dict = {"x": 10, "y": 30, "z": 20}
max_key = max(my_dict, key=my_dict.get)
print(max_key)  # 결과: y
```
